# Óptica geométrica

Teoría:
- Ecuacion de la lente delgada:

    $ \Phi = \frac{n'}{s'} - \frac{n}{s} = \frac{n'}{f'} = - \frac{n}{f} $,
    
    donde si $\Phi > 0$ es una lente convergente y caso contrario será divergente.


- Parametros de interés:
$\begin{itemize}
    1- $S_{objeto}$: Posición del objeto

    2- $S_{imagen}$: Posición en que se forma la imagen

    3- $n_1$: Índice de refracción del medio de incidencia

    4- $n_2$: Índice de refracción del medio de salida

    5- $h_1$: Altura original del objeto

    6- $h_2$: Altura de la imagen

    6- (i) El signo nos indica si la img es directa o invertida
     
    6- (ii) El modulo nos indica si es aumentada o disminuida


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Posiciones (metros)
S_obj: float = -10.
S_img: float = ...

# Indices de refraccion
n_1: float = 1.
n_2: float = 1.33 # agua

# Alturas (metros)
h_1: float = 0.1 
h_2: float = ...

# Focos (metros)
f_obj: float = -5.
f_img: float = ...

# Potencia de la lente
phi: float = - n_1 / f_obj

# Calculo lo que falta

S_img: float = n_2 / (phi + n_1 / S_obj)

h_2:   float = h_1 * n_1 * S_img / (n_2 * S_obj)

f_img: float = n_2 / phi
```

Con estos datos, y lo calculado a partir de ellos, necesito calcular la *pendiente* de las rectas (S, h)->(f, 0) y (0, h)->(f', 0), para poder graficarlas y extenderlas hasta \\ (0, h') y (S', h') respectivamente

Llamo $m_1$ a la primera y $m_2$ a la segunda

```python
m_1: float = -h_1 / (f_obj - S_obj)
m_2: float = -h_1 / (f_img)
```

```python
plt.figure(figsize=(12,6))

# Flecha indicando punta del objeto
plt.arrow(
    S_obj, 0, dx = 0, dy = h_1,
    length_includes_head = True,
    color = 'blue', alpha = .75,
    linewidth = 2, head_length = 0.2
)

# Flecha horizontal hasta el eje 'y'
plt.arrow(
    S_obj, h_1,
    dx = -S_obj, dy = 0,
    length_includes_head = True,
    color = 'red', alpha = .75,
    linewidth = 2, head_length = 0.2
)

# Flecha de (0, h) hasta (s', h')
plt.arrow(
    0, h_1, dx = S_img, dy = - h_1 + h_2,
    length_includes_head = True,
    color = 'red', alpha = .75,
    linewidth = 2, head_length = 0.2
)

# Flecha desde (s, h) hasta (0, h')
plt.arrow(
    S_obj, h_1, dx = -S_obj, dy = - h_1 + h_2,
    length_includes_head = True,
    color = 'violet', alpha = .75,
    linewidth = 2, head_length = 0.2
)

# Flecha horizontal hasta (s', h')
plt.arrow(
    0, h_2, dx = S_img, dy = 0,
    length_includes_head = True,
    color = 'violet', alpha = .75,
    linewidth = 2, head_length = 0.2
)

# Flecha vertical en la imagen
plt.arrow(
    S_img, 0, dx = 0, dy = h_2,
    length_includes_head = True,
    color = 'blue', alpha = .75,
    linewidth = 2, head_length = 0.2
)

plt.hlines(y=0, xmin=-15, xmax=15,
           linestyles='dashed',
           alpha = 0.75, colors='black')

plt.vlines(x=[S_obj, f_obj, f_img, 0],
           ymin = h_2,
           ymax = [0, h_1, h_1, h_1],
           colors = 'black',
           linestyles='dashed',
           alpha=.5
           )

plt.xticks([S_obj, f_obj, 0, f_img, S_img], 
           ['$S_{objeto}$', '$f_{objeto}$',
            0, 
            '$f_{imagen}$', '$S_{imagen}$'])

plt.show()
```
